```python
import pandas as pd
import numpy as np
import panel as pn
import matplotlib.pyplot as plt
pn.extension('tabulator')
import datetime as dt

import hvplot.pandas
```

```python
df = pd.read_csv('P6-UK-Bank-Customers.csv')
```

```python
df
```

```python
df.columns
```

```python
unique_formats = df['Date Joined'].unique()

print("Nilai unik dalam kolom 'Date Joined':")
print(unique_formats)
```

```python
def parse_date_safe(date_str):
    try:
        return pd.to_datetime(date_str, format='%d.%b.%y')
    except:
        return None
```

```python
df['Date Joined'] = df['Date Joined'].apply(parse_date_safe)
```

```python
start_date = pd.Timestamp('2015-01-05')
end_date = pd.Timestamp('2015-12-30')
df = df[(df['Date Joined'] >= start_date) & (df['Date Joined'] <= end_date)]
```

```python
if df['Date Joined'].isnull().sum() > 0:
    print("Ada nilai kosong yang ditemukan setelah filter.")
else:
    print("Semua nilai dalam kolom 'Date Joined' valid.")
```

```python
idf = df.interactive()
```

```python
date_range_slider = pn.widgets.DateRangeSlider(
    name='Date Joined',
    start=dt.datetime(2015, 1, 5), end=dt.datetime(2015, 12, 30),
    value=(dt.datetime(2015, 2, 5), dt.datetime(2015, 11, 30)),
    step=2
)

date_range_slider
```

```python
# Radio button untuk perhitungan CO2
Balance_radio = pn.widgets.RadioButtonGroup(
    name='Y Balance', 
    options=['Balance', 'Age',],
    button_type='primary'
)
Balance_radio
```

```python
unique_formats = df['Region'].unique()
print("Nilai unik dalam kolom 'Region':")
print(unique_formats)
```

```python
# Menghubungkan data pipeline dan widget yang telah dibuat
continents = ['England', 'Northern Ireland', 'Wales', 'Scotland']

start_date = pd.to_datetime(start_date)
end_date = pd.to_datetime(end_date) 

Balance_pipeline = (
    idf[
        (idf['Date Joined'] >= start_date) & (idf['Date Joined'] <= end_date) &
            (df.Region.isin(continents))
        ]
        .groupby(['Region', 'Date Joined'])[Balance_radio].mean()
        .to_frame()
        .reset_index()
        .sort_values(by='Date Joined')  
        .reset_index(drop=True)
    )
Balance_pipeline
```

```python
Balance_plot = Balance_pipeline.hvplot(x = 'Date Joined', by='Region', y=Balance_radio,line_width=2, title="Saldo Pelanggan Menurut Negara nya")
Balance_plot
```

```python
Balance_table = Balance_pipeline.pipe(pn.widgets.Tabulator, pagination='remote', page_size = 10, sizing_mode='stretch_width') 
Balance_table
```

```python
Job_source_bar_pipeline = (
   idf
    .groupby('Job Classification')['Customer ID']
    .sum()
    .reset_index(drop=False)  # Menyimpan index hasil groupby
)
```

```python
Job_source_bar_plot = Job_source_bar_pipeline.hvplot(kind='bar', 
                                                     x='Job Classification', 
                                                     y='Customer ID', 
                                                     title='Jumlah Pelanggan Berdasarkan Klasifikasi Pekerjaan')
Job_source_bar_plot
```

```python
#Layout using Template
template = pn.template.FastListTemplate(
    title='Dashboard Data Customer Bank', 
    sidebar=[pn.pane.Markdown("# Customers Bank"), 
             pn.pane.Markdown("#### Pengelolaan data customers yang tepat memastikan bahwa semua data disimpan dan diproses sesuai dengan regulasi yang berlaku. ."), 
             pn.pane.Markdown("## Settings"),   
             date_range_slider],
    main=[pn.Row(pn.Column(Balance_radio, 
                           Balance_plot.panel(width=700), margin=(0,25)), 
                 Balance_table.panel(width=500)), 
          pn.Row(pn.Column(Job_source_bar_plot.panel(width=1200), margin=(0,5)))], 
    accent_base_color="#42468e",
    header_background="#42468e",
)
# template.show()
template.servable();
```
